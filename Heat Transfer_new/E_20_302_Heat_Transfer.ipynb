# **Cooling Load Calculation**

**Given Data**

\begin{align*}
\text{Number of containers} &= 40 \\
m_{\text{milk}} &= 5 \ \text{kg (per container)} \\
m_{\text{container}} &= 2 \ \text{kg (per container)} \\
c_{\text{Al}} &= 921.1 \ \text{J/kg·K} \\
c_{\text{milk}} &= 4000 \ \text{J/kg·K} \\
T_i &= 40^\circ \text{C} \\
T_f &= 15^\circ \text{C} \\
t &= 2 \ \text{hours} = 7200 \ \text{s}
\end{align*}

Temperature Drop

\begin{align*}
\Delta T &= T_i - T_f \\
&= 40 - 15 \\
&= 25^\circ \text{C}
\end{align*}

**Heat Removed from Milk**

For one container:

\begin{align*}
Q_{\text{milk,1}} &= m c \Delta T \\
&= (5)(4000)(25) \\
&= 500{,}000 \ \text{J}
\end{align*}

For 40 containers:

\begin{align*}
Q_{\text{milk,total}} &= 40 \times 500{,}000 \\
&= 20{,}000{,}000 \ \text{J}
\end{align*}

**Heat Removed from Aluminum Containers**

For one container:

\begin{align*}
Q_{\text{container,1}} &= (2)(921.1)(25) \\
&= 46{,}055 \ \text{J}
\end{align*}

For 40 containers:

\begin{align*}
Q_{\text{container,total}} &= 40 \times 46{,}055 \\
&= 1{,}842{,}200 \ \text{J}
\end{align*}

**Total Heat Load**

\begin{align*}
Q_{\text{total}} &= Q_{\text{milk,total}} + Q_{\text{container,total}} \\
&= 20{,}000{,}000 + 1{,}842{,}200 \\
&= 21{,}842{,}200 \ \text{J} \\
&\approx 21.84 \ \text{MJ}
\end{align*}

**Required Cooling Power**

\begin{align*}
P &= \frac{Q_{\text{total}}}{t} \\
&= \frac{21{,}842{,}200}{7200} \\
&\approx 3033.6 \ \text{W} \\
&\approx 3.03 \ \text{kW}
\end{align*}

To determine the feasibility of the available air conditioning unit, a structured engineering evaluation method is applied.

**Proposed Method to Determine Feasibility**

1. Calculation of Required Cooling Load

The first step is to determine the total heat that must be removed from the system. This includes:

*   Heat removed from milk
*   Heat removed from aluminum containers
*   Additional heat gains (room heat leakage, door opening, lighting, etc.)


From previous calculations:
\begin{align*}
Required cooling Capacity\approx 3.03 \ \text{kW}
\end{align*}

This represents the design cooling load.

2. Determination of Available AC Unit Capacity


*   Cooling capacity (kW or BTU/hr)
*   Input power (kW)
*   Refrigerant type
*   Coefficient of Performance (COP)

Then compare:
\begin{align*}
 $\text{Available Capacity} \geq \text{Required Cooling Load}$
\end{align*}



Feasibility condition:

| Condition                           | Conclusion       |
| ----------------------------------- | ---------------- |
| Capacity < 3 kW                     | Not feasible     |
| Capacity ≈ 3 kW                     | Marginal         |
| Capacity ≥ 3.5 kW                   | Feasible         |
