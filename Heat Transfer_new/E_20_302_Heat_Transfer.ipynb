# **Cooling Load Calculation**

**Given Data**

\begin{align*}
\text{Number of containers} &= 40 \\
m_{\text{milk}} &= 5 \ \text{kg (per container)} \\
m_{\text{container}} &= 2 \ \text{kg (per container)} \\
c_{\text{Al}} &= 921.1 \ \text{J/kg·K} \\
c_{\text{milk}} &= 4000 \ \text{J/kg·K} \\
T_i &= 40^\circ \text{C} \\
T_f &= 15^\circ \text{C} \\
t &= 2 \ \text{hours} = 7200 \ \text{s}
\end{align*}

Temperature Drop

\begin{align*}
\Delta T &= T_i - T_f \\
&= 40 - 15 \\
&= 25^\circ \text{C}
\end{align*}

**Heat Removed from Milk**

For one container:

\begin{align*}
Q_{\text{milk,1}} &= m c \Delta T \\
&= (5)(4000)(25) \\
&= 500{,}000 \ \text{J}
\end{align*}

For 40 containers:

\begin{align*}
Q_{\text{milk,total}} &= 40 \times 500{,}000 \\
&= 20{,}000{,}000 \ \text{J}
\end{align*}

**Heat Removed from Aluminum Containers**

For one container:

\begin{align*}
Q_{\text{container,1}} &= (2)(921.1)(25) \\
&= 46{,}055 \ \text{J}
\end{align*}

For 40 containers:

\begin{align*}
Q_{\text{container,total}} &= 40 \times 46{,}055 \\
&= 1{,}842{,}200 \ \text{J}
\end{align*}

**Total Heat Load**

\begin{align*}
Q_{\text{total}} &= Q_{\text{milk,total}} + Q_{\text{container,total}} \\
&= 20{,}000{,}000 + 1{,}842{,}200 \\
&= 21{,}842{,}200 \ \text{J} \\
&\approx 21.84 \ \text{MJ}
\end{align*}

**Required Cooling Power**

\begin{align*}
P &= \frac{Q_{\text{total}}}{t} \\
&= \frac{21{,}842{,}200}{7200} \\
&\approx 3033.6 \ \text{W} \\
&\approx 3.03 \ \text{kW}
\end{align*}

# **Feasibility Assessment of the Available Air Conditioning Unit**

To determine the feasibility of the available air conditioning unit, a structured engineering evaluation method is applied.

**Proposed Method to Determine Feasibility**

1. Calculation of Required Cooling Load

The first step is to determine the total heat that must be removed from the system. This includes:

*   Heat removed from milk
*   Heat removed from aluminum containers
*   Additional heat gains (room heat leakage, door opening, lighting, etc.)


From previous calculations:
\begin{align*}
Required cooling Capacity\approx 3.03 \ \text{kW}
\end{align*}

This represents the design cooling load.

2. Determination of Available AC Unit Capacity


*   Cooling capacity (kW or BTU/hr)
*   Input power (kW)
*   Refrigerant type
*   Coefficient of Performance (COP)

Then compare:
\begin{align*}
 $\text{Available Capacity} \geq \text{Required Cooling Load}$
\end{align*}



Feasibility condition:

| Condition                           | Conclusion       |
| ----------------------------------- | ---------------- |
| Capacity < 3 kW                     | Not feasible     |
| Capacity ≈ 3 kW                     | Marginal         |
| Capacity ≥ 3.5 kW                   | Feasible         |

3. Perform an Experimental Test

Experimental Procedure:

I. Place thermal load inside the room

Measure:

Initial temperature

Final temperature after 2 hours

Room air temperature


II. Calculate actual heat removed:

\begin{align*}
Q = m c_p (T_i - T_f)
\end{align*}

III. Determine actual cooling rate
\begin{align*}
\dot Q = Q/t
\end{align*}

Compare this value with the required 3 kW
4. Evaluate Performance Criteria

*  Milk temperature reduces to 15°C within 2 hours
*  Compressor does not run continuously (no overload)
*  Room temperature remains stable
*  Unit operates safely without frosting or pressure issues

5. Consider Safety Margin

For industrial applications, include a 15--25\% safety factor:
\begin{align*}
\
3\,\text{kW} \times 1.2 = 3.6\,\text{kW}
\
\end{align*}
So the recommended practical capacity is:
\begin{align*}
\approx 3.5\text{-}4\,\text{kW}
\
\end{align*}

```python
import numpy as np

# Atmospheric pressure (kPa)
P = 101.325

# Air flow rate (kg/hr)
air_flow_hr = 30

# Convert to kg/s
m_dot = air_flow_hr / 3600

# Supply air temperatures (DB and WB)
Tdb_supply = np.array([24.1,23.3,22.7,22.4,22.1,21.9,21.8,21.6,21.4,21.2,21.1,21.0,20.9])
Twb_supply = np.array([20.6,20.4,19.8,19.3,18.9,18.7,18.4,18.2,18.0,17.8,17.7,17.6,17.4])

# Return air temperatures (DB and WB)
Tdb_return = np.array([17.2,17.1,16.7,16.6,16.5,16.4,16.2,16.1,15.9,15.9,15.7,15.7,15.4])
Twb_return = np.array([17.1,17.1,16.9,16.8,16.6,16.5,16.4,16.3,16.2,16.1,16.0,16.0,15.8])

# Average temperatures
Tdb_s = np.mean(Tdb_supply)
Twb_s = np.mean(Twb_supply)

Tdb_r = np.mean(Tdb_return)
Twb_r = np.mean(Twb_return)

# Saturation vapor pressure function (kPa)
def Psat(T):
    return 0.61078 * np.exp((17.27*T)/(T+237.3))

# Humidity ratio calculation
def humidity_ratio(Tdb, Twb):
    Pws = Psat(Twb)
    A = 0.00066 * (1 + 0.00115 * Twb)
    Pw = Pws - A * P * (Tdb - Twb)
    w = 0.62198 * Pw / (P - Pw)
    return w

# Calculate humidity ratios
w_s = humidity_ratio(Tdb_s, Twb_s)
w_r = humidity_ratio(Tdb_r, Twb_r)

# Enthalpy calculation (kJ/kg)
def enthalpy(T, w):
    return 1.005*T + w*(2501 + 1.88*T)

h_s = enthalpy(Tdb_s, w_s)
h_r = enthalpy(Tdb_r, w_r)

# Cooling capacity
Q = m_dot * (h_s - h_r)   # kW because enthalpy in kJ/kg

# Print results
print("Average Supply Air DB:", round(Tdb_s,2),"°C")
print("Average Return Air DB:", round(Tdb_r,2),"°C")

print("Supply Air Enthalpy:", round(h_s,2),"kJ/kg")
print("Return Air Enthalpy:", round(h_r,2),"kJ/kg")

print("Cooling Capacity:", round(Q,4),"kW")

# Required cooling load
Q_required = 3.03

if Q >= Q_required:
    print("AC unit is FEASIBLE")
else:
    print("AC unit is NOT FEASIBLE")
```
