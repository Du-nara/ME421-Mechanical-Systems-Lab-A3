# Part 1: Method to Evaluate the Feasibility of the Available AC Unit

## 1. Objective

The objective of this analysis is to determine whether the Air Conditioning unit available in the Heat Engine Laboratory can:

1. Cool 40 milk containers from 40°C to 15°C within 2 hours.
2. Maintain the milk at 15°C for approximately 24 hours.

The feasibility study consists of an analytical cooling load calculation followed by an experimental validation.

---

## 2. Given Data

$$
\begin{aligned}
\text{Number of containers} &= 40 \\
m_{\text{milk}} &= 5 \text{ kg per container} \\
m_{\text{Al}} &= 2 \text{ kg per container} \\
c_{\text{milk}} &= 4000 \text{ J/kg K} \\
c_{\text{Al}} &= 921.1 \text{ J/kg K} \\
T_i &= 40^\circ C \\
T_f &= 15^\circ C \\
t &= 2 \text{ hours} = 7200 \text{ s}
\end{aligned}
$$

---

## 3. Temperature Reduction Required

$$
\Delta T = T_i - T_f
$$

$$
\Delta T = 40 - 15 = 25 \text{ K}
$$

This represents the temperature difference through which both milk and container material must be cooled.

---

## 4. Heat Removal from Milk

The cooling energy required for one container of milk:

$$
Q_{\text{milk,1}} = m c \Delta T
$$

$$
Q_{\text{milk,1}} = (5)(4000)(25)
$$

$$
Q_{\text{milk,1}} = 500000 \text{ J}
$$

For 40 containers:

$$
Q_{\text{milk,total}} = 40 \times 500000
$$

$$
Q_{\text{milk,total}} = 20000000 \text{ J}
$$

Engineering Note:  
Milk has a high specific heat capacity, therefore it represents the major portion of the cooling load.

---

## 5. Heat Removal from Aluminum Containers

For one aluminum container:

$$
Q_{\text{Al,1}} = (2)(921.1)(25)
$$

$$
Q_{\text{Al,1}} = 46055 \text{ J}
$$

For 40 containers:

$$
Q_{\text{Al,total}} = 40 \times 46055
$$

$$
Q_{\text{Al,total}} = 1842200 \text{ J}
$$

Although aluminum has lower specific heat than milk, its contribution must be included for accurate estimation.

---

## 6. Total Cooling Load

$$
Q_{\text{total}} = Q_{\text{milk,total}} + Q_{\text{Al,total}}
$$

$$
Q_{\text{total}} = 20000000 + 1842200
$$

$$
Q_{\text{total}} = 21842200 \text{ J}
$$

$$
Q_{\text{total}} \approx 21.84 \text{ MJ}
$$

This is the total thermal energy that must be removed within 2 hours.

---

## 7. Required Cooling Power

Cooling power is obtained by dividing total energy by time:

$$
\dot{Q} = \frac{Q_{\text{total}}}{t}
$$

$$
\dot{Q} = \frac{21842200}{7200}
$$

$$
\dot{Q} \approx 3034 \text{ W}
$$

$$
\dot{Q} \approx 3.03 \text{ kW}
$$

This value represents the theoretical minimum refrigeration capacity required.

---

## 8. Inclusion of Practical Safety Margin

In real applications, additional heat gains occur due to:

- Heat transfer through walls
- Door opening
- Air infiltration
- Internal fan heat
- Measurement uncertainties

Assuming a 30 percent safety factor:

$$
Q_{\text{design}} = 3.03 \times 1.3
$$

$$
Q_{\text{design}} \approx 3.94 \text{ kW}
$$

Therefore, a practical refrigeration capacity of approximately 4 kW is recommended.

---

## 9. Feasibility Comparison Criteria

After obtaining the AC unit nameplate data:

- Cooling capacity
- Input power
- COP

Feasibility condition:

If

$$
\text{AC Capacity} \geq 4 \text{ kW}
$$

The unit is suitable.

If capacity is close to 3 kW, performance will be marginal and may fail during peak load conditions.

---

## 10. Maintenance Load Evaluation

Once milk reaches 15°C, the AC only needs to offset heat gain from surroundings.

Maintenance load can be estimated using:

$$
Q = UA (T_{\text{ambient}} - T_{\text{room}})
$$

Where:

- U = Overall heat transfer coefficient  
- A = Surface area  
- Temperature difference = Ambient minus storage temperature  

If this maintenance load is lower than AC capacity, temperature stability can be maintained.

---

## 11. Experimental Verification Plan

To validate calculations, perform a controlled laboratory test:

1. Place equivalent thermal mass (water containers) inside the room.
2. Measure:
   - Initial product temperature
   - Final temperature after 2 hours
   - Room air temperature variation
   - Compressor operating cycle
3. Calculate actual cooling rate using:

$$
Q = mc(T_i - T_f)
$$

$$
\dot{Q} = Q/t
$$

4. Compare experimental cooling rate with required 3 kW.

---

## 12. Final Decision Criteria

The AC unit is considered feasible if:

- Milk temperature reduces to 15°C within 2 hours
- Required cooling rate is achieved
- Compressor operates without continuous overload
- Stable 15°C temperature is maintained

---

## Conclusion

The analytical study shows that approximately 3.03 kW is required theoretically, while about 4 kW is recommended under practical conditions.  

The final feasibility decision must be confirmed through experimental validation of the actual AC unit performance.
