```python
import sympy as sp

# Define symbols
t = sp.symbols('t')
theta = sp.Function('theta')(t)
M_b, M_m, L, L1, k, m0, e, omega = sp.symbols('M_b M_m L L_1 k m_0 e omega')

# Moment of Inertia about the pivot
J0 = (1/3)*M_b*L**2 + M_m*L1**2

# Equation of Motion: J0*theta'' + k*L^2*theta = m0*e*omega^2*L1*sin(omega*t)
lhs = J0 * theta.diff(t, t) + k * L**2 * theta
rhs = m0 * e * omega**2 * L1 * sp.sin(omega * t)

EOM = sp.Eq(lhs, rhs)
print("The derived Equation of Motion is:")
display(EOM)
```

```python
import sympy as sp
from IPython.display import display

# 1. Define symbolic variables
t = sp.symbols('t')
theta = sp.Function('theta')(t)
Mb, Mm, L, L1, k, m0, e, omega = sp.symbols('M_b M_m L L_1 k m_0 e omega')

# 2. Total Moment of Inertia about pivot O
Jo = (1/3)*Mb*L**2 + Mm*L1**2

# 3. Define the Equation of Motion (EOM)
# Jo * theta_ddot + k_eq * theta = Torque_amplitude * sin(omega * t)
lhs = Jo * theta.diff(t, t) + (k * L**2) * theta
rhs = (m0 * e * L1) * omega**2 * sp.sin(omega * t)

EOM = sp.Eq(lhs, rhs)

print("Mathematical Model for Activity #1:")
display(EOM)

# 4. Calculate Natural Frequency (omega_n)
wn = sp.sqrt((k * L**2) / Jo)
print("\nTheoretical Natural Frequency (rad/s):")
display(wn)
```

Generalized Force (Excitation)The unbalanced motor creates a harmonic force $F(t) = m_0 e \omega^2 \sin(\omega t)$, where $m_0$ is the unbalanced mass and $e$ is the eccentricity. Since it acts at distance $L_1$, the torque is:$$M(t) = F(t) \cdot L_1 = (m_0 e \omega^2 \sin(\omega t)) L_1$$4. Equation of MotionUsing the Lagrange or Newton’s Second Law for rotation ($\sum M_0 = J_0 \alpha$):$$J_0 \ddot{\theta} + k L^2 \theta = m_0 e \omega^2 L_1 \sin(\omega t)$$

```python
import sympy as sp
from IPython.display import display

# Define symbols
Mb, Mm, L, L1, k, Ct = sp.symbols('M_b M_m L L_1 k C_t')

# 1. Define Jo (Moment of Inertia)
Jo = (1/3)*Mb*L**2 + Mm*L1**2

# 2. Define k_eq (Equivalent Torsional Stiffness)
k_eq = k * L**2

# 3. Define Symbolic Natural Frequency (omega_n)
omega_n = sp.sqrt(k_eq / Jo)

# 4. Define Critical Damping (Cc) and Damping Ratio (zeta)
Cc = 2 * sp.sqrt(k_eq * Jo)
zeta = Ct / Cc

print("Activity #2: Symbolic Expressions")
print("-" * 30)
print("Equivalent Stiffness (k_eq):")
display(k_eq)

print("\nUndamped Natural Frequency (omega_n):")
display(omega_n)

print("\nDamping Ratio (zeta):")
display(zeta)
```

From Activity #1, we have:$$J_O \ddot{\theta} + k L^2 \theta = \tau(t)$$To find the natural frequency, we look at the unforced, undamped version:$$J_O \ddot{\theta} + (k L^2) \theta = 0$$Divide by $J_O$ to get the standard form $\ddot{\theta} + \omega_n^2 \theta = 0$:$$\ddot{\theta} + \left( \frac{k L^2}{J_O} \right) \theta = 0$$Step 2: Symbolic Expression for Natural Frequency ($\omega_n$)The term multiplying $\theta$ is defined as the square of the undamped natural frequency.$$\omega_n = \sqrt{\frac{k L^2}{J_O}}$$Substituting the expression for $J_O$ (Inertia of beam + motor):$$\omega_n = \sqrt{\frac{k L^2}{\frac{1}{3}M_b L^2 + M_m L_1^2}}$$Step 3: Including Damping (Symbolic $\zeta$)If the lab requires you to account for damping (from the pivot or air resistance), we add a damping term $C_t \dot{\theta}$. The EOM becomes:$$J_O \ddot{\theta} + C_t \dot{\theta} + k L^2 \theta = \tau(t)$$The symbolic Damping Ratio ($\zeta$) is expressed as:$$\zeta = \frac{C_t}{C_c} = \frac{C_t}{2 \sqrt{k L^2 J_O}}$$

Activity 3
Step 1: The Particular SolutionFor a harmonic excitation $M(t) = (m_0 e L_1 \omega^2) \sin(\omega t)$, we assume a steady-state solution of the form:$$\theta(t) = \Theta \sin(\omega t - \phi)$$Where:$\Theta$ is the Amplitude of angular vibration.$\phi$ is the Phase Angle (the lag between the force and the movement).Step 2: Symbolic Expression for Amplitude ($\Theta$)Using the standard solution for a forced 1-DOF system:$$\Theta = \frac{M_0}{\sqrt{(k_{eq} - J_O \omega^2)^2 + (C_t \omega)^2}}$$Substituting our specific terms from previous activities:$M_0 = m_0 e L_1 \omega^2$ (Excitation Torque)$k_{eq} = k L^2$$J_O = \frac{1}{3} M_b L^2 + M_m L_1^2$The expression becomes:$$\Theta(\omega) = \frac{m_0 e L_1 \omega^2}{\sqrt{(k L^2 - J_O \omega^2)^2 + (C_t \omega)^2}}$$Step 3: Non-Dimensional Form (Frequency Response Function)To make this useful for simulations, we often express the amplitude in terms of the frequency ratio $r = \frac{\omega}{\omega_n}$:$$\Theta = \frac{(m_0 e L_1 / J_O) r^2}{\sqrt{(1 - r^2)^2 + (2 \zeta r)^2}}$$This formula allows you to see how the system behaves at resonance ($r \approx 1$).

```python
import sympy as sp
from IPython.display import display

# Define symbols
t, omega = sp.symbols('t omega')
Mb, Mm, L, L1, k, m0, e, Ct = sp.symbols('M_b M_m L L_1 k m_0 e C_t')

# Parameters from Activity 1 & 2
Jo = (1/3)*Mb*L**2 + Mm*L1**2
keq = k * L**2
M0 = m0 * e * L1 * omega**2  # Force Amplitude

# Define the Steady-State Amplitude (Theta)
# Using the formula: M0 / sqrt((keq - Jo*omega**2)**2 + (Ct*omega)**2)
denominator = sp.sqrt((keq - Jo * omega**2)**2 + (Ct * omega)**2)
Theta = M0 / denominator

print("Activity #3: Steady-State Response Analysis")
print("-" * 45)
print("Symbolic Amplitude (Theta) as a function of omega:")
display(Theta)

# Define the Phase Angle (phi)
phi = sp.atan2(Ct * omega, keq - Jo * omega**2)
print("\nSymbolic Phase Angle (phi):")
display(phi)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Physical Parameters (Standard for TecQuipment) ---
Mb = 1.35      # Beam mass (kg)
Mm = 1.2       # Motor mass (kg)
L = 0.711      # Total length (m)
L1 = 0.355     # Motor distance (m)
k = 1000       # Spring stiffness (N/m)
m0_e = 0.0005  # Unbalance product (kg·m)
zeta = 0.05    # Damping ratio (5%)

# --- 2. Calculations ---
Jo = (1/3)*Mb*L**2 + Mm*L1**2
keq = k * L**2
wn = np.sqrt(keq / Jo)
fn = wn / (2 * np.pi)

# Frequency range for plotting (from 0 to 15 Hz)
f_hz = np.linspace(0.1, 15, 1000)
omega = 2 * np.pi * f_hz
r = omega / wn  # Frequency ratio

# --- 3. Compute the Three Responses ---

# Plot 1: Displacement Amplitude (X_ss at the end of the beam)
# Formula: M0 / sqrt((keq - Jo*w^2)^2 + (Ct*w)^2)
M0 = m0_e * L1 * omega**2
denominator = np.sqrt((keq - Jo * omega**2)**2 + (2 * zeta * Jo * wn * omega)**2)
Theta = M0 / denominator
X_ss = L * Theta * 1000 # Convert to mm

# Plot 2: Phase Angle (Phi)
# Formula: atan( (2*zeta*r) / (1 - r^2) )
phase = np.arctan2(2 * zeta * r, 1 - r**2)
phase_deg = np.degrees(phase)

# Plot 3: Transmissibility (TR)
# TR = sqrt( 1 + (2*zeta*r)^2 ) / sqrt( (1-r^2)^2 + (2*zeta*r)^2 )
TR = np.sqrt(1 + (2*zeta*r)**2) / np.sqrt((1 - r**2)**2 + (2*zeta*r)**2)

# --- 4. Plotting ---
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Subplot 1: Amplitude
axs[0].plot(f_hz, X_ss, color='blue', lw=2)
axs[0].axvline(fn, color='red', linestyle='--', label=f'Resonance ({fn:.2f} Hz)')
axs[0].set_title('1. Displacement Amplitude Response')
axs[0].set_ylabel('Amplitude (mm)')
axs[0].grid(True)
axs[0].legend()

# Subplot 2: Phase Angle
axs[1].plot(f_hz, phase_deg, color='green', lw=2)
axs[1].axvline(fn, color='red', linestyle='--')
axs[1].set_title('2. Phase Angle Response')
axs[1].set_ylabel('Phase (Degrees)')
axs[1].set_yticks([0, 45, 90, 135, 180])
axs[1].grid(True)

# Subplot 3: Transmissibility
axs[2].plot(f_hz, TR, color='purple', lw=2)
axs[2].axvline(fn, color='red', linestyle='--')
axs[2].axhline(1, color='black', linestyle='-')
axs[2].set_title('3. Displacement Transmissibility')
axs[2].set_xlabel('Frequency (Hz)')
axs[2].set_ylabel('TR Ratio')
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
