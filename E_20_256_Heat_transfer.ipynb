**Part 1: Method to Evaluate the Feasibility of the Available AC Unit**

**1. Objective**

The objective of this analysis is to determine whether the Air Conditioning unit available in the Heat Engine Laboratory can:

1)Cool 40 milk containers from 40°C to 15°C within 2 hours.

2)Maintain the milk at 15°C for approximately 24 hours.

```python
# ── GIVEN DATA ────────────────────────────────────────────────────────────────
# You can change these values if your data is different

n_containers   = 40       # Number of milk containers
m_milk         = 5        # Mass of milk per container (kg)
m_Al           = 2        # Mass of aluminium container (kg)
c_milk         = 4000     # Specific heat of milk (J/kg·K)
c_Al           = 921.1    # Specific heat of aluminium (J/kg·K)
T_initial      = 40       # Initial temperature (°C)  ← worst case
T_final        = 15       # Target temperature (°C)
t_cooling      = 2 * 3600 # Cooling time in seconds (2 hours = 7200 s)

print('Given Data Summary')
print('='*45)
print(f'  Number of containers       : {n_containers}')
print(f'  Mass of milk / container   : {m_milk} kg')
print(f'  Mass of Al  / container    : {m_Al} kg')
print(f'  Specific heat of milk      : {c_milk} J/kg·K')
print(f'  Specific heat of aluminium : {c_Al} J/kg·K')
print(f'  Initial temperature        : {T_initial} °C')
print(f'  Target temperature         : {T_final} °C')
print(f'  Cooling time               : {t_cooling} s  (2 hours)')
```
